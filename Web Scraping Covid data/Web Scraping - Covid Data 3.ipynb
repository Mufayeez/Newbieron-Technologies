```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_html('https://www.statista.com/statistics/1103458/india-novel-coronavirus-covid-19-cases-by-state/')
```

```python
data[0]
```

```python
data[0].to_csv('corona_cases_state_wise.csv')
```

```python
corona = pd.read_csv('corona_cases_state_wise.csv')
```

```python
corona.head()
```

```python
corona.tail()
```

```python
corona = corona.drop(labels=['Unnamed: 0'], axis=1)
corona
```

```python
states = corona.sort_values('Deceased', ascending=False).head(5)
plt.pie(states['Deceased'], labels = states['Characteristic'], autopct = '%1.1f%%')
plt.title('Top 5 State Death Rate')
plt.show()
```
