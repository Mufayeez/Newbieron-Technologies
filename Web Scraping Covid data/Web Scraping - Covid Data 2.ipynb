```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_html("https://prsindia.org/covid-19/cases")
```

```python
data[0].to_csv("corona_cases_state_wise.csv")
```

```python
corona_cases = pd.read_csv('corona_cases_state_wise.csv')
```

```python
corona_cases = corona_cases.drop(labels=['Unnamed: 0', '#'], axis=1)
```

```python
corona_cases = corona_cases.drop(36, inplace=False)
```

```python
corona_cases.head()
```

```python
most_deaths = corona_cases.sort_values('Death', ascending=False).head(5)
sns.barplot(x='State/UT', y ='Death', data=most_deaths)
```

```python
most_active = corona_cases.sort_values('Active Cases', ascending=False).head(5)
sns.barplot(x = "State/UT", y = 'Active Cases', data=most_active)
```

```python
confirmed_cases = corona_cases.sort_values('Confirmed Cases', ascending=False).head(5)
sns.barplot(x = 'State/UT', y = 'Confirmed Cases', data = confirmed_cases)
```

```python
cured = corona_cases.sort_values('Cured/Discharged', ascending=False).head(5)
sns.barplot(x = 'State/UT', y = 'Cured/Discharged', data = cured)
```
